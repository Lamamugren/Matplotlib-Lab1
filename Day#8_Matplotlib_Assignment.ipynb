## Q5: Create three line charts with Legends at left and colors

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [2, 8, 18, 32, 50]
y3 = [3, 12, 27, 48, 75]

plt.plot(x, y1, label="programming", color='red')
plt.plot(x, y2, label="leadership", color='green')
plt.plot(x, y3, label="Public Speaking", color='blue')
plt.legend(loc='upper left')
plt.show()
```

## Q6: Create subplots with three charts and figsize=20,10

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [2, 8, 18, 32, 50]
y3 = [3, 12, 27, 48, 75]

fig, axs = plt.subplots(1, 3, figsize=(20, 10))
axs[0].plot(x, y1)
axs[0].set_title('programming Cours')
axs[1].plot(x, y2)
axs[1].set_title('leadership Cours ')
axs[2].plot(x, y3)
axs[2].set_title('Public Speaking Cours ')
plt.show()
```
